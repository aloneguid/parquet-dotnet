# Reading DataFrame with Parquet.Net

This sample uses [Parquet.Net](https://github.com/aloneguid/parquet-dotnet) with it's [built-in support](https://github.com/aloneguid/parquet-dotnet/blob/master/docs/dataframe.md) for [DataFrame API](https://github.com/aloneguid/parquet-dotnet/blob/master/docs/dataframe.md).

In the next cell, we install Parquet.Net from nuget.org and import required namespaces.

```polyglot-notebook
#r "nuget: Parquet.Net"

using System;
using System.IO;
using System.Net.Http;
using Parquet;
using Microsoft.Data.Analysis;
```

Now, download a parquet file. In this sample, I download a test file from parquet.net repository. Then create a stream `ms` from it.

```polyglot-notebook
var hc = new HttpClient();
byte[] fileData = await hc.GetByteArrayAsync("https://raw.githubusercontent.com/aloneguid/parquet-dotnet/master/src/Parquet.Test/data/postcodes.plain.parquet");
var ms = new MemoryStream(fileData);
```

Using `System.IO.Stream`'s extension method `ReadParquetAsDataFrameAsync` read the stream to DataFrame in just one line of code

```polyglot-notebook
DataFrame df = await ms.ReadParquetAsDataFrameAsync();
```

Now visualise it, as DataFrame has support for Jupyter already
